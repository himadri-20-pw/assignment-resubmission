**Question 1:  What is K-Nearest Neighbors (KNN) and how does it work in both classification and regression problems?** 

**Answer:** 

K-Nearest Neighbhors(KNN) is a non-parametric,instance-based learning algorithm used for both classification and regression tasks by analyzing the proximity of data points.It also called as a lazy learner alogrithm because here we are calculating parameters of model in real time. It works at real time the distance of test data from each of datapoint is calculated, then training and computation time is more, that’s why it is computationally expensive. KNN is easy to understand/very intuitive and the performance of model in terms of evaluation matric is good.KNN is a supervised machine learning algorithm generally used for classification but can also be used for regression tasks.For a new data point, KNN identifies its k closest neighbors from the training data and then predicts the class by a majority vote (classification) or the average value (regression) of these neighbors.Since KNN makes no assumptions about the underlying data distribution it makes it a non-parametric,instance-based learning method.

**How KNN Works:-**

1.Choose the value of 'k': 
The number of nearest neighbors to consider, represented by 'k', is a user-defined constant. 

2.Calculate distances: 
For a new, unlabeled data point, the algorithm calculates the distance between it and every other data point in the training set. 

3.Find the k nearest neighbors: 
The 'k' data points with the smallest distances are identified as the nearest neighbors. 

4.Make a prediction:

-->For classification: The algorithm assigns the new data point the class that is most frequently present among its k nearest neighbors (a "majority vote"). 

-->For regression: The algorithm predicts the new data point's value as the average of the values of its k nearest neighbors.

**Question 2: What is the Curse of Dimensionality and how does it affect KNN 
performance?** 

**Answer:** 

The Curse of Dimensionality describes how increasing the number of features in a dataset makes it exponentially harder to work with, as the number of dimensions increases, the data becomes increasingly sparse, and the distance between data points becomes less meaningful. It leading to issues like data sparsity and distorted distance metrics.For the K-Nearest Neighbors (KNN) algorithm, this curse causes performance to degrade in high dimensions because data points become equally distant from each other, making it difficult to find genuinely "nearest" neighbors and increasing the need for exponentially more data to avoid overfitting.

**How it affects KNN performance:**

KNN relies on the distance between data points to classify new data. The Curse of Dimensionality negatively impacts KNN in the following ways: 

• Loss of Discrimination: 
In high dimensions, all points tend to be equidistant from each other. This makes it difficult for KNN to find the true "nearest" neighbors, reducing its ability to correctly classify new data points. 

• Increased sparsity:
 In high dimensions, data points become very far apart, making it difficult to find truly "nearest" neighbors. Most points will appear equally distant, reducing the effectiveness of distance-based metrics. 

• Increased Data Requirements:
To represent patterns effectively in high-dimensional space, KNN needs a very large dataset.
Otherwise, predictions may be based on noise rather than meaningful neighbors.

• Overfitting:
 In high-dimensional spaces, if there isn’t enough data to cover the feature space, KNN may overfit. Instead of learning generalizable patterns, the model can memorize noise or detect spurious correlations between features and target variables. This leads to poor performance on unseen data.

• Algorithmic Inefficiency: 
KNN requires distance computation between a test point and all training points.In high dimensions, this becomes computationally expensive and slows prediction.
 
• Reduced interpretability:
High-dimensional data is hard to visualize.Relationships between features and neighbors are difficult to understand intuitively.

**Question 3: What is Principal Component Analysis (PCA)? How is it different from feature selection?** 

**Answer:**

Principal Component Analysis (PCA) is a dimensionality reduction technique used in machine learning and statistics. It transforms a set of possibly correlated variables into a set of linearly uncorrelated variables called principal components. The first principal component accounts for the largest possible variance in the data, and each succeeding component accounts for the largest possible remaining variance.

**Difference :**

--->Principal Component Analysis (PCA):-

• Purpose: To reduce data dimensionality by transforming high-dimensional, correlated data into a smaller set of uncorrelated principal components that capture most of the original data's variance. 

• Method: It uses linear algebra to find new axes (principal components) that align with the directions of the greatest data variance. 

• Outcome: New, linearly combined features (principal components) that are ranked by their ability to explain the variance in the dataset. 

• Nature: Primarily an unsupervised learning and feature extraction technique, not a feature selection method itself.

--->Feature Selection:-

• Purpose: To reduce the number of features to improve model training speed, reduce the chance of overfitting, and gain insights into which original features are most important for predicting a target variable.

• Method: It ranks and then selects a subset of the most relevant original features, discarding the less important ones.

• Outcome: A smaller, more manageable set of the original features.

• Nature: Often used in supervised learning to identify features most useful for predicting a target variable.

**Key Differences:**

1.Transformation vs. Selection: 
PCA creates new features (components) by transforming the original ones, while feature selection chooses a subset of the existing, original features.

2.Feature Nature:
 PCA's new components are linear combinations of the original features, whereas feature selection keeps the original features intact.

3.Goal: 
PCA's goal is to reduce dimensionality while preserving variance and removing redundancy, while feature selection's goal is to identify the most informative original features for a specific task.
4.Interpretability: 
PCA components can be difficult to interpret in terms of the original meaning of the features, while feature selection allows for clear interpretation of the selected features' importance.

**Question 4: What are eigenvalues and eigenvectors in PCA, and why are they 
important?** 

**Answer:**

Eigenvalues and Eigenvectors are fundamental concepts in Principal Component Analysis (PCA) used for dimensionality reduction and feature extraction. 

• Eigenvectors: These represent the principal components or directions of maximum variance in the data. They are orthogonal to each other and indicate the directions along which the data is most spread out. When the original data is transformed by the covariance matrix, an eigenvector's direction remains unchanged, only its magnitude may be scaled by its eigenvalue.

• Eigenvalues: Each eigenvalue is a scalar value that represents the amount of variance captured by its corresponding eigenvector. Larger eigenvalues correspond to eigenvectors that explain more of the data's total variability, while smaller eigenvalues indicate directions with less variation.Eigenvalues quantify the information content of each principal component.

**Importance in PCA:**

Eigenvalues and eigenvectors are crucial in PCA for the following reasons: 

1.Dimensionality Reduction:
By selecting the eigenvectors with the largest eigenvalues, we can reduce the dimensionality of the data while retaining the most important information (variance). We discard principal components with small eigenvalues as they represent directions with less variance and thus carry less significant information. 

2.Feature Extraction: 
The eigenvectors (principal components) become the new features or dimensions of the transformed data. These new features are uncorrelated, which can be beneficial for subsequent machine learning tasks. 

3.Data Visualization:
The principal components can be used to visualize high-dimensional data in lower dimensions, making patterns and relationships in the data more apparent. 

**Question 5: How do KNN and PCA complement each other when applied in a single 
pipeline?** 

**Answer:**
When K-Nearest Neighbors (KNN) and Principal Component Analysis (PCA) are used together in a single pipeline, they complement each other by addressing different aspects of data processing and model performance: 

• Dimensionality Reduction (PCA): 
PCA is used as a preprocessing step to reduce the dimensionality of the dataset by transforming the original features into a smaller set of uncorrelated principle components that capture most of the variance in the data. This is particularly beneficial for KNN because KNN is sensitive to the curse of dimensionality, where high-dimensional data can lead to increased computational cost and reduced accuracy due to sparse data points and noisy features. By reducing the number of features, PCA can make KNN more efficient and potentially more accurate by focusing on the most important underlying patterns in the data. 

• Noise Reduction (PCA): 
PCA can help in noise reduction by transforming the data into a new set of principal components, where the components capturing the most variance are retained, and those representing noise are often discarded. This can lead to a cleaner dataset for KNN, improving its ability to find accurate nearest neighbors. 

• Improved Performance (KNN): 
With the reduced and potentially cleaner dataset from PCA, KNN can perform better. The reduced dimensionality means that distance calculations are faster, and the nearest neighbors are more likely to be truly representative of the data's underlying structure, leading to improved classification or regression performance for KNN. 

In essence, PCA acts as a powerful preprocessing step for KNN, optimizing the data for KNN's distance-based classification or regression.KNN then leverages this optimized, lower-dimensional representation to find the nearest neighbors more efficiently and accurately, leading to better overall model performance.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
pca
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
print(f"Original number of features: {X.shape[1]}")
print(f"Number of features after PCA: {X_train_pca.shape[1]}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train_pca,y_train)
```

```python
y_pred = knn.predict(X_test_pca)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=cancer.target_names))
```

-->Impact:
We keep the most informative signals while discarding irrelevant variability — making downstream analysis faster, more reliable, and interpretable.

-->Impact:

-->Impact: 
The pipeline is scientifically rigorous, statistically validated, and minimizes overfitting.
